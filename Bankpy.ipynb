```python
menu = """
   üè¶
[1] Depositar
[2] Sacar
[3] Extrato
[4] Sair

=> """

saldo = 0
limite = 500
extrato = ""
numero_saques = 0
LIMITE_SAQUES = 3

while True:

    opcao = input(menu)

    if opcao == "1":
        valor = float(input("üí≤Informe o valor do dep√≥sito: "))

        if valor > 0:
            saldo += valor
            extrato += f"Dep√≥sito: R$ {valor:.2f}\n"

        else:
            print("‚ùå Opera√ß√£o falhou! O valor informado √© inv√°lido.")

    elif opcao == "2":
        valor = float(input("üí∏Informe o valor do saque: "))

        excedeu_saldo = valor > saldo

        excedeu_limite = valor > limite

        excedeu_saques = numero_saques >= LIMITE_SAQUES

        if excedeu_saldo:
            print("Opera√ß√£o falhou! Voc√™ n√£o tem saldo suficiente.")

        elif excedeu_limite:
            print("‚ùå Opera√ß√£o falhou! O valor do saque excede o limite.")

        elif excedeu_saques:
            print("‚ùå Opera√ß√£o falhou! N√∫mero m√°ximo de saques excedido.")

        elif valor > 0:
            saldo -= valor
            extrato += f"Saque: R$ {valor:.2f}üí∏\n"
            numero_saques += 1

        else:
            print("‚ùå Opera√ß√£o falhou! O valor informado √© inv√°lido.")

    elif opcao == "3":
        print("\n================ EXTRATO ================")
        print("N√£o foram realizadas movimenta√ß√µes." if not extrato else extrato)
        print(f"\nSaldo: R$ {saldo:.2f}üí∞")
        print("==========================================")

    elif opcao == "4":
        break

    else:
        print("Opera√ß√£o inv√°lida, por favor selecione novamente a opera√ß√£o desejada.")
```
